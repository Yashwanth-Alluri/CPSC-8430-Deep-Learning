```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
from torch.autograd import Variable

torch.manual_seed(42)

train_dataset = torchvision.datasets.MNIST(root='./data', train=True, transform=transforms.Compose([transforms.ToTensor()]), download=True)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, transform=transforms.Compose([transforms.ToTensor()]))

random_labels = torch.tensor(np.random.randint(0, 10, (len(train_dataset),)))
train_dataset.targets = random_labels

train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=128, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=128, shuffle=False)

class RandomMNIST(nn.Module):
    def __init__(self):
        super(RandomMNIST, self).__init__()
        self.fc1 = nn.Linear(784, 256)
        self.fc2 = nn.Linear(256, 256)
        self.fc3 = nn.Linear(256, 256)
        self.out = nn.Linear(256, 10)

    def forward(self, x):
        x = x.view(-1, 784)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = F.relu(self.fc3(x))
        x = self.out(x)
        return x

def train_model(model, optimizer, epoch, train_loader):
    model.train()
    loss_fn = nn.CrossEntropyLoss()
    running_loss = 0.0
    for i, (data, target) in enumerate(train_loader):
        data, target = Variable(data), Variable(target)
        optimizer.zero_grad()
        outputs = model(data)
        loss = loss_fn(outputs, target)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()
        if (i+1) % 100 == 0: 
            print(f'Epoch [{epoch+1}], Step [{i+1}/{len(train_loader)}], Loss: {loss.item():.4f}')

    avg_train_loss = running_loss / len(train_loader)
    return avg_train_loss

def test_model(model, test_loader):
    model.eval()
    loss_fn = nn.CrossEntropyLoss()
    total_loss = 0
    with torch.no_grad():
        for i, (data, target) in enumerate(test_loader):
            data, target = Variable(data), Variable(target)
            outputs = model(data)
            loss = loss_fn(outputs, target)
            total_loss += loss.item()

        avg_test_loss = total_loss / len(test_loader)
        print(f'Test Loss: {avg_test_loss:.4f}')
    
    return avg_test_loss

model = RandomMNIST()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0003)
train_loss_arr, test_loss_arr = [], []
epochs = 500
for epoch in range(epochs):
    print(f'--- Epoch {epoch+1}/{epochs} ---')
    train_loss = train_model(model, optimizer, epoch, train_loader)
    test_loss = test_model(model, test_loader)
    
    print(f'Epoch [{epoch+1}] | Train Loss: {train_loss:.4f} | Test Loss: {test_loss:.4f}\n')
    
    train_loss_arr.append(train_loss)
    test_loss_arr.append(test_loss)
```

```python
plt.plot(range(epochs), train_loss_arr, color="green")
plt.plot(range(epochs), test_loss_arr, color="purple")
plt.legend(['Train Loss', 'Test Loss'], loc='upper left')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Random Labels Train vs Test Loss', color='darkblue')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
from torch.autograd import Variable
import copy

torch.manual_seed(42)

train_dataset = torchvision.datasets.MNIST(root='./data', train=True, transform=transforms.ToTensor(), download=True)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, transform=transforms.ToTensor())

train_loader_64 = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=64, shuffle=True)
train_loader_1024 = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=1024, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=1024, shuffle=False)

class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.fc1 = nn.Linear(784, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 10)

    def forward(self, x):
        x = x.view(-1, 784)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

def train_model(model, optimizer, train_loader, num_epochs):
    model.train()
    loss_fn = nn.CrossEntropyLoss()
    for epoch in range(num_epochs):
        total_loss, correct, total = 0, 0, 0
        for images, labels in train_loader:
            images, labels = Variable(images), Variable(labels)
            optimizer.zero_grad()
            outputs = model(images)
            loss = loss_fn(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        avg_loss = total_loss / len(train_loader)
        acc = 100.0 * correct / total
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}, Accuracy: {acc:.2f}%")
    
    return model

def test_model(model, test_loader):
    model.eval()
    loss_fn = nn.CrossEntropyLoss()
    correct, total, total_loss = 0, 0, 0
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = Variable(images), Variable(labels)
            outputs = model(images)
            loss = loss_fn(outputs, labels)
            total_loss += loss.item()

            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    avg_loss = total_loss / len(test_loader)
    accuracy = 100.0 * correct / total
    return avg_loss, accuracy

def interpolate_models(alpha, theta_1, theta_2):
    return (1 - alpha) * theta_1 + alpha * theta_2

def plot_results(alphas, train_loss, test_loss, train_acc, test_acc, lr_label):
    fig, ax1 = plt.subplots()

    ax1.set_xlabel('alpha')
    ax1.set_ylabel('cross_entropy in log', color='blue')
    ax1.plot(alphas, np.log(train_loss), 'b--', label=f'Train ({lr_label})')
    ax1.plot(alphas, np.log(test_loss), 'b-', label=f'Test ({lr_label})')
    ax1.tick_params(axis='y', labelcolor='blue')

    ax2 = ax1.twinx()
    ax2.set_ylabel('accuracy', color='red')
    ax2.plot(alphas, train_acc, 'r--', label=f'Train ({lr_label})')
    ax2.plot(alphas, test_acc, 'r-', label=f'Test ({lr_label})')
    ax2.tick_params(axis='y', labelcolor='red') 

    fig.tight_layout()
    plt.title(f'Linear Interpolation between Models ({lr_label})')
    plt.legend()
    plt.show()

m1_lr1 = SimpleModel()
optimizer_m1_lr1 = torch.optim.Adam(m1_lr1.parameters(), lr=1e-2)
m1_trained_lr1 = train_model(m1_lr1, optimizer_m1_lr1, train_loader_64, num_epochs=5)

m2_lr1 = SimpleModel()
optimizer_m2_lr1 = torch.optim.Adam(m2_lr1.parameters(), lr=1e-2)
m2_trained_lr1 = train_model(m2_lr1, optimizer_m2_lr1, train_loader_1024, num_epochs=5)

theta_1_lr1 = torch.nn.utils.parameters_to_vector(m1_trained_lr1.parameters()).detach()
theta_2_lr1 = torch.nn.utils.parameters_to_vector(m2_trained_lr1.parameters()).detach()

alphas = np.linspace(-1.0, 2.0, 30)
train_loss_lr1_model1, test_loss_lr1_model1 = [], []
train_loss_lr1_model2, test_loss_lr1_model2 = [], []
train_acc_lr1_model1, test_acc_lr1_model1 = [], []
train_acc_lr1_model2, test_acc_lr1_model2 = [],[]

for alpha in alphas:
    theta_alpha_lr1 = interpolate_models(alpha, theta_1_lr1, theta_2_lr1)
    
    interpolated_model_lr1 = SimpleModel()
    torch.nn.utils.vector_to_parameters(theta_alpha_lr1, interpolated_model_lr1.parameters())

    train_loss_1, train_acc_1 = test_model(interpolated_model_lr1, train_loader_64)
    test_loss_1, test_acc_1 = test_model(interpolated_model_lr1, test_loader)

    train_loss_lr1_model1.append(train_loss_1)
    test_loss_lr1_model1.append(test_loss_1)
    train_acc_lr1_model1.append(train_acc_1)
    test_acc_lr1_model1.append(test_acc_1)

plot_results(alphas, train_loss_lr1_model1, test_loss_lr1_model1, train_acc_lr1_model1, test_acc_lr1_model1, "lr1 = 1e-2")


m1_lr2 = SimpleModel()
optimizer_m1_lr2 = torch.optim.Adam(m1_lr2.parameters(), lr=1e-3)
m1_trained_lr2 = train_model(m1_lr2, optimizer_m1_lr2, train_loader_64, num_epochs=5)

m2_lr2 = SimpleModel()
optimizer_m2_lr2 = torch.optim.Adam(m2_lr2.parameters(), lr=1e-3)
m2_trained_lr2 = train_model(m2_lr2, optimizer_m2_lr2, train_loader_1024, num_epochs=5)

theta_1_lr2 = torch.nn.utils.parameters_to_vector(m1_trained_lr2.parameters()).detach()
theta_2_lr2 = torch.nn.utils.parameters_to_vector(m2_trained_lr2.parameters()).detach()

train_loss_lr2_model1, test_loss_lr2_model1 = [], []
train_loss_lr2_model2, test_loss_lr2_model2 = [],[]
train_acc_lr2_model1, test_acc_lr2_model1 = [], []
train_acc_lr2_model2, test_acc_lr2_model2 = [],[]

for alpha in alphas:
    theta_alpha_lr2 = interpolate_models(alpha, theta_1_lr2, theta_2_lr2)
    
    interpolated_model_lr2 = SimpleModel()
    torch.nn.utils.vector_to_parameters(theta_alpha_lr2, interpolated_model_lr2.parameters())

    train_loss_2, train_acc_2 = test_model(interpolated_model_lr2, train_loader_64)
    test_loss_2, test_acc_2 = test_model(interpolated_model_lr2, test_loader)

    train_loss_lr2_model1.append(train_loss_2)
    test_loss_lr2_model1.append(test_loss_2)
    train_acc_lr2_model1.append(train_acc_2)
    test_acc_lr2_model1.append(test_acc_2)

plot_results(alphas, train_loss_lr2_model1, test_loss_lr2_model1, train_acc_lr2_model1, test_acc_lr2_model1, "lr2 = 1e-3")
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
from torch.autograd import Variable

torch.manual_seed(42)

train_dataset = torchvision.datasets.MNIST(root='./data', train=True, transform=transforms.ToTensor(), download=True)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False, transform=transforms.ToTensor())

batch_size_values = [64, 128, 256, 512, 1024]  

class CustomModel(nn.Module):
    def __init__(self):
        super(CustomModel, self).__init__()
        self.fc1 = nn.Linear(784, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 10)

    def forward(self, x):
        x = x.view(-1, 784)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

def train_model(custom_model, opt, data_loader, epochs):
    custom_model.train()
    criterion = nn.CrossEntropyLoss()
    for epoch in range(epochs):
        cumulative_loss, correct_preds, total_samples = 0, 0, 0
        for images, labels in data_loader:
            images, labels = Variable(images), Variable(labels)
            opt.zero_grad()
            outputs = custom_model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            opt.step()

            cumulative_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_samples += labels.size(0)
            correct_preds += (predicted == labels).sum().item()

        avg_loss = cumulative_loss / len(data_loader)
        accuracy = 100.0 * correct_preds / total_samples
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {avg_loss:.4f}, Accuracy: {accuracy:.2f}%")
    
    return avg_loss, accuracy

def test_model(custom_model, test_data_loader):
    custom_model.eval()
    criterion = nn.CrossEntropyLoss()
    correct_preds, total_samples, total_loss = 0, 0, 0
    with torch.no_grad():
        for images, labels in test_data_loader:
            images, labels = Variable(images), Variable(labels)
            outputs = custom_model(images)
            loss = criterion(outputs, labels)
            total_loss += loss.item()

            _, predicted = torch.max(outputs.data, 1)
            total_samples += labels.size(0)
            correct_preds += (predicted == labels).sum().item()

    avg_loss = total_loss / len(test_data_loader)
    accuracy = 100.0 * correct_preds / total_samples
    return avg_loss, accuracy

train_loss_list, test_loss_list = [], []
train_acc_list, test_acc_list = [], []
sensitivity_list = []

for b_size in batch_size_values:
    print(f"\nTraining with batch size: {b_size}")
    
    train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=b_size, shuffle=True)
    test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=b_size, shuffle=False)
    
    custom_model = CustomModel()
    optimizer = torch.optim.Adam(custom_model.parameters(), lr=0.001)
    
    train_loss, train_acc = train_model(custom_model, optimizer, train_loader, epochs=20)
    
    test_loss, test_acc = test_model(custom_model, test_loader)
    
    train_loss_list.append(train_loss)
    test_loss_list.append(test_loss)
    train_acc_list.append(train_acc)
    test_acc_list.append(test_acc)
    
    sensitivity = abs(train_loss - test_loss) / (train_loss + test_loss)
    sensitivity_list.append(sensitivity)

```

```python
fig_loss_sensitivity, loss_ax = plt.subplots()

loss_ax.set_xlabel('Batch Size (log scale)')
loss_ax.set_xscale('log')
loss_ax.set_ylabel('Cross Entropy / Loss', color='purple')
loss_ax.plot(batch_size_values, train_loss_list, 'g-', label='Train Loss', linewidth=2.0)
loss_ax.plot(batch_size_values, test_loss_list, 'g--', label='Test Loss', linewidth=2.0)
loss_ax.tick_params(axis='y', labelcolor='purple')

sensitivity_ax = loss_ax.twinx()
sensitivity_ax.set_ylabel('Sensitivity', color='red')
sensitivity_ax.plot(batch_size_values, sensitivity_list, 'r-', label='Sensitivity', linewidth=2.0)
sensitivity_ax.tick_params(axis='y', labelcolor='red')

fig_loss_sensitivity.tight_layout()
loss_ax.legend(loc='upper left')
sensitivity_ax.legend(loc='upper right')
plt.title('Cross Entropy Loss and Sensitivity vs Batch Size (log scale)')
plt.show()


fig_accuracy_sensitivity, accuracy_ax = plt.subplots()

accuracy_ax.set_xlabel('Batch Size (log scale)')
accuracy_ax.set_xscale('log')
accuracy_ax.set_ylabel('Accuracy (%)', color='blue')
accuracy_ax.plot(batch_size_values, train_acc_list, 'b-', label='Train Accuracy', linewidth=2.0)
accuracy_ax.plot(batch_size_values, test_acc_list, 'b--', label='Test Accuracy', linewidth=2.0)
accuracy_ax.tick_params(axis='y', labelcolor='blue')

sensitivity_ax2 = accuracy_ax.twinx()
sensitivity_ax2.set_ylabel('Sensitivity', color='red')
sensitivity_ax2.plot(batch_size_values, sensitivity_list, 'r-', label='Sensitivity', linewidth=2.0)
sensitivity_ax2.tick_params(axis='y', labelcolor='red')


fig_accuracy_sensitivity.tight_layout()
accuracy_ax.legend(loc='upper left')
sensitivity_ax2.legend(loc='upper right')
plt.title('Accuracy and Sensitivity vs Batch Size (log scale)')
plt.show()
```
