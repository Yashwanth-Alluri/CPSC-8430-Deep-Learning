```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt

def function1(x):
    return np.sinc(5 * np.pi * x / np.pi) 

def function2(x):
    return np.sign(np.sin(5 * np.pi * x))

def generate_training_data(func):
    x_vals = np.linspace(-1, 1, 1000)
    y_vals = func(x_vals)
    x_tensor = torch.tensor(x_vals, dtype=torch.float32).view(-1, 1)
    y_tensor = torch.tensor(y_vals, dtype=torch.float32).view(-1, 1)
    return x_tensor, y_tensor

def plot_functions():
    x_vals = np.linspace(-1, 1, 1000)

    y_vals_func1 = function1(x_vals)
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(x_vals, y_vals_func1, label="Function 1: sinc(5*pi*x)")
    plt.title("Function 1: sinc(5*pi*x) / (5*pi*x)")
    plt.xlabel("x")
    plt.ylabel("y")

    y_vals_func2 = function2(x_vals)
    plt.subplot(1, 2, 2)
    plt.plot(x_vals, y_vals_func2, label="Function 2: sgn(sin(5*pi*x))", color="red")
    plt.title("Function 2: sgn(sin(5*pi*x))")
    plt.xlabel("x")
    plt.ylabel("y")

    plt.tight_layout()
    plt.show()

# Model 1 (Shallow)
class Model1(nn.Module):
    def __init__(self):
        super(Model1, self).__init__()
        self.fc1 = nn.Linear(1, 5)
        self.fc2 = nn.Linear(5, 10)
        self.fc3 = nn.Linear(10, 10)
        self.fc4 = nn.Linear(10, 10)
        self.fc5 = nn.Linear(10, 10)
        self.fc6 = nn.Linear(10, 5)
        self.fc7 = nn.Linear(5, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = torch.relu(self.fc4(x))
        x = torch.relu(self.fc5(x))
        x = torch.relu(self.fc6(x))
        return self.fc7(x)

# Model 2 (Moderate)
class Model2(nn.Module):
    def __init__(self):
        super(Model2, self).__init__()
        self.fc1 = nn.Linear(1, 10)
        self.fc2 = nn.Linear(10, 18)
        self.fc3 = nn.Linear(18, 15)
        self.fc4 = nn.Linear(15, 4)
        self.fc5 = nn.Linear(4, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = torch.relu(self.fc4(x))
        return self.fc5(x)

# Model 3 (Deep)
class Model3(nn.Module):
    def __init__(self):
        super(Model3, self).__init__()
        self.fc1 = nn.Linear(1, 190)
        self.fc2 = nn.Linear(190, 1)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        return self.fc2(x)

def train_model(model, input_data, output_data, epochs=20000, learning_rate=0.001):
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    loss_function = nn.MSELoss()
    losses = []

    for epoch in range(epochs):
        optimizer.zero_grad()
        predictions = model(input_data)
        loss = loss_function(predictions, output_data)
        loss.backward()
        optimizer.step()
        losses.append(loss.item())
        
        if epoch % 1000 == 0:
            print(f'Epoch {epoch}, Loss: {loss.item():.4f}')

    return losses

def plot_loss_comparison(losses1, losses2, losses3, title):
    plt.figure(figsize=(8, 6))
    plt.plot(losses1, label="Model 1 (Shallow)", color="blue")
    plt.plot(losses2, label="Model 2 (Moderate)", color="green")
    plt.plot(losses3, label="Model 3 (Deep)", color="red")
    plt.title(f'Loss Comparison - {title}')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

def plot_prediction_comparison(model1, model2, model3, x_data, y_data, title):
    model1.eval()
    model2.eval()
    model3.eval()
    with torch.no_grad():
        predictions1 = model1(x_data).detach().numpy()
        predictions2 = model2(x_data).detach().numpy()
        predictions3 = model3(x_data).detach().numpy()

    plt.figure(figsize=(8, 6))
    plt.plot(x_data.numpy(), y_data.numpy(), label="True Function", color="black")
    plt.plot(x_data.numpy(), predictions1, label="Model 1 (Shallow)", color="blue")
    plt.plot(x_data.numpy(), predictions2, label="Model 2 (Moderate)", color="green")
    plt.plot(x_data.numpy(), predictions3, label="Model 3 (Deep)", color="red")
    plt.title(f'Prediction Comparison - {title}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    plot_functions()

    x_train1, y_train1 = generate_training_data(function1)
    
    model1 = Model1()
    print("Training Model 1 on Function 1...")
    model1_losses = train_model(model1, x_train1, y_train1)

    model2 = Model2()
    print("Training Model 2 on Function 1...")
    model2_losses = train_model(model2, x_train1, y_train1)

    model3 = Model3()
    print("Training Model 3 on Function 1...")
    model3_losses = train_model(model3, x_train1, y_train1)

    plot_loss_comparison(model1_losses, model2_losses, model3_losses, "Function 1")

    plot_prediction_comparison(model1, model2, model3, x_train1, y_train1, "Function 1")
    
    x_train2, y_train2 = generate_training_data(function2)

    print("Training Model 1 on Function 2...")
    model1_losses = train_model(model1, x_train2, y_train2)

    print("Training Model 2 on Function 2...")
    model2_losses = train_model(model2, x_train2, y_train2)

    print("Training Model 3 on Function 2...")
    model3_losses = train_model(model3, x_train2, y_train2)

    plot_loss_comparison(model1_losses, model2_losses, model3_losses, "Function 2")

    plot_prediction_comparison(model1, model2, model3, x_train2, y_train2, "Function 2")
```
